Questão 3 - Mapa do Brasil

Crie um sistema para representar o mapa do Brasil, através dos seus estados. Esse sistema,
além de representar o nosso país, também deverá exibir os vizinhos diretos de cada estado.
Por exemplo, SP possui PR, MS, MG e RJ como estados vizinhos diretos. Por fim, permita que
o sistema receba dois parâmetros, um estado de origem e um de destino, e retorne um
caminho possível para ir do estado de origem para o estado de destino.

```python
class Vertex: # classe para representar um vértice do grafo
    def __init__(self, label, acronyms): # construtor da classe, recebe um rótulo e uma sigla para o vértice
        self.label = label
        self.acronyms = acronyms
        self.neighbors = set()
        
    def __str__(self): # método para representar o vértice como uma string, mostrando seu rótulo, sigla e vizinhos
        return str(self.label) + ' (' + str(self.acronyms) + ')' + ' neighbors: ' + str([str(x.label) for x in self.neighbors])
    
class Graph:
    def __init__(self): # construtor da classe, inicializa um dicionário para armazenar os vértices do grafo
        self.vertices = {}
        
    def add_vertex(self, vertex): # método para adicionar um vértice ao grafo, usando seu rótulo como chave no dicionário
        self.vertices[vertex.label] = vertex # adiciona o vértice ao dicionário, associando seu rótulo ao objeto vértice
        
    def add_edge(self, from_vertex, to_vertex):
        self.vertices[from_vertex.label].neighbors.add(to_vertex) # adiciona o vértice de destino como vizinho do vértice de origem
        self.vertices[to_vertex.label].neighbors.add(from_vertex) # adiciona o vértice de origem como vizinho do vértice de destino, já que o grafo é não direcionado
        
    def print_graph(self):
        for vertex in self.vertices:
            print(self.vertices[vertex])
```

```python
def bfs(graph, start, end): # função para realizar a busca em largura (BFS) no grafo, recebendo o grafo, o vértice de início e o vértice de fim como parâmetros
    visited = set()
    queue = []
    path = []
    
    queue.append([start]) # inicia a fila com o caminho contendo apenas o vértice de início
    visited.add(start) # marca o vértice de início como visitado
    
    while queue:
        path = queue.pop(0) # remove o primeiro caminho da fila para explorar
        current = path[-1] # obtém o último vértice do caminho atual para verificar se é o vértice de destino

        if current == end: # verifica se o vértice atual é o vértice de destino
            print('Caminho encontrado!')
            return path

        for neighbor in graph.vertices[current].neighbors:
            if neighbor.label not in visited: # verifica se o vizinho ainda não foi visitado
                visited.add(neighbor.label) # marca o vizinho como visitado para evitar ciclos
                new_path = path + [neighbor.label] # cria um novo caminho adicionando o vizinho ao caminho atual
                queue.append(new_path)
                
       # print(
        #    f"Actual: {current} | Path: {path} | Queue: {queue} | Visited: {sorted(visited)}"
        #)
```

```python
# dict para armazenar os estados do Brasil, associando suas siglas aos seus nomes completos
list_states_brazil = {
    "AC": "Acre - AC",
    "AL": "Alagoas - AL",
    "AP": "Amapa - AP",
    "AM": "Amazonas - AM",
    "BA": "Bahia - BA",
    "CE": "Ceara - CE",
    "DF": "Distrito Federal - DF",
    "ES": "Espirito Santo - ES",
    "GO": "Goias - GO",
    "MA": "Maranhao - MA",
    "MT": "Mato Grosso - MT",
    "MS": "Mato Grosso do Sul - MS",
    "MG": "Minas Gerais - MG",
    "PA": "Para - PA",
    "PB": "Paraiba - PB",
    "PR": "Parana - PR",
    "PE": "Pernambuco - PE",
    "PI": "Piaui - PI",
    "RJ": "Rio de Janeiro - RJ",
    "RN": "Rio Grande do Norte - RN",
    "RS": "Rio Grande do Sul - RS",
    "RO": "Rondonia - RO",
    "RR": "Roraima - RR",
    "SC": "Santa Catarina - SC",
    "SP": "Sao Paulo - SP",
    "SE": "Sergipe - SE",
    "TO": "Tocantins - TO",
}

neighbors = {
    "AC": ["AM", "RO"],
    "AL": ["BA", "PE", "SE"],
    "AP": ["PA"],
    "AM": ["AC", "RO", "MT", "PA", "RR"],
    "BA": ["AL", "ES", "GO", "MG", "PE", "PI", "SE", "TO"],
    "CE": ["PI", "RN", "PB", "PE"],
    "DF": ["GO", "MG"],
    "ES": ["BA", "MG", "RJ"],
    "GO": ["DF", "BA", "MG", "MS", "MT", "TO"],
    "MA": ["PA", "PI", "TO"],
    "MT": ["AM", "PA", "RO", "MS", "GO", "TO"],
    "MS": ["MT", "GO", "MG", "SP", "PR"],
    "MG": ["BA", "ES", "GO", "DF", "MS", "SP", "RJ"],
    "PA": ["AP", "AM", "RR", "MT", "TO", "MA"],
    "PB": ["RN", "CE", "PE"],
    "PR": ["MS", "SP", "SC"],
    "PE": ["AL", "BA", "PI", "CE", "PB"],
    "PI": ["MA", "TO", "BA", "PE", "CE"],
    "RJ": ["ES", "MG", "SP"],
    "RN": ["CE", "PB"],
    "RS": ["SC"],
    "RO": ["AC", "AM", "MT"],
    "RR": ["AM", "PA"],
    "SC": ["PR", "RS"],
    "SP": ["MS", "MG", "RJ", "PR"],
    "SE": ["AL", "BA"],
    "TO": ["PA", "MA", "PI", "BA", "GO", "MT"],
}
```

```python
vertex = {acronyms: Vertex(acronyms, nome) for acronyms, nome in list_states_brazil.items()}

graph = Graph()
for v in vertex.values():
    graph.add_vertex(v)

for origin, destinos in neighbors.items():
    for dest in destinos:
        graph.add_edge(vertex[origin], vertex[dest])

for key, neighbors in neighbors.items():
    print(f"{key} está conectado com → {', '.join(neighbors)}")

origin = input("Adicione a origem: ")
destination = input("Adicione o destino: ")
origin = origin.strip()
destination = destination.strip()

print(f"Buscando caminho de {origin} para {destination}...")

for key in list_states_brazil.keys():
    if origin not in list_states_brazil:
        print(f"Origem '{origin}' não encontrada. Verifique a sigla e tente novamente.")
        exit()
    if destination not in list_states_brazil:
        print(f"Destino '{destination}' não encontrado. Verifique a sigla e tente novamente.")
        exit()

path = bfs(graph, origin.upper(), destination.upper())
print(" → ".join(path))
```

### COMPLEXIDADE DO ALGORITMO 

Grafo + BFS: O(V+E), linear no tamanho do grafo (vertices + arestas) porque o BFS visita cada vertice no maixmo uma vez 
e ao fazer isso varre todas as arestas ligadas a esses vertices.




Questão 5 - Sugestão de Arquivos 
 
Supondo  que  vocês  estão  implementando  uma  solução  que  poderá  ser  utilizada  por 
qualquer  tipo  de  aplicação  para  recomendar  qual  a  pasta  do  computador  com  mais 
arquivos relacionados. Por exemplo, supondo que estamos utilizando uma aplicação de 
edição de imagens, a solução irá sugerir a pasta do computador que mais possui imagens. 
Se estivermos utilizando uma aplicação para reproduzir músicas, a solução irá sugerir a 
pasta que mais possui arquivos de áudio. 
 
Considerando um exemplo de estrutura de arquivos de um determinado computador: 
home

│   ├── documentos

│   │   └── tcc.docx 
│   ├── imagens 
│   │   ├── foto1.jpg 
│   │   └── foto2.jpg 
│   └── musicas 
│       └── rock.mp3 
└── admin 
    └── configs 
        └── system.cfg 
 
 
A tarefa de vocês é: 
• Identificar e implementar qual estrutura melhor represente problema; 
• Implementar uma maneira de recomendar uma pasta (path completo da pasta) que 
possui o maior número de arquivos relacionados dado um tipo de arquivo informado 
pelo usuário; 
• Exibir  a  listagem  de  arquivos  e  o  caminho  que  foi  percorrido  para  chegar  nesses 
arquivos.

```python
# criando uma classe que representa uma árvore
class Tree:
    # inicializando a árvore com um nodo
    def __init__(self, node):
        self.node = node # representa o nodo da árvore
        self.children = [] # lista de filhos do nodo

    # adicionando um filho ao nodo
    def add_child(self, child):
        # adicionando o filho à lista de filhos
        self.children.append(child)

    # representação da árvore em string
    def __str__(self, level=0):
        ret = ("\t" * level) + str(self.node) + "\n"
        for child in self.children:
            ret += child.__str__(level + 1)
        return ret
```

```python
# função para realizar a busca em profundidade (DFS) na árvore, recebendo o nodo raiz, 
# a extensão de arquivo a ser buscada, uma lista para armazenar os resultados e um dicionário para 
# contar a quantidade de arquivos encontrados em cada pasta
def dfs(root, ext, count):
    if root is None:
        return

    # percorrendo os filhos do nodo atual
    for child in root.children: # para cada filho do nodo atual, verificamos se o nome do filho termina com a extensão desejada
        if child.name.endswith(ext):
            pasta = "/".join(node.name for node in child.path[:-1]) # obtém o caminho da pasta onde o arquivo está localizado, juntando os nomes dos nodos do caminho, exceto o último (que é o arquivo)
            count[pasta] = count.get(pasta, 0) + 1 # contando a quantidade de arquivos com a extensão em cada pasta
            print(f"Arquivo encontrado: {child.name} | Pasta: {pasta} | Contagem atual: {count[pasta]}")

        dfs(child, ext, count) 
```

```python
structure = [
    ("user", "root"),
    ("admin", "root"),
    ("documents", "user"),
    ("pictures", "user"),
    ("musics", "user"),
    ("videos", "user"),

    ("spreadsheets.xls", "documents"),

    ("projects", "documents"),
    ("reports", "documents"),

    ("photos_2024", "pictures"),
    ("trip", "photos_2024"),

    ("rock", "musics"),
    ("pop", "musics"),

    ("movies", "videos"),

    ("pdfs", "admin"),
    ("configs", "admin"),

    ("report.docx", "reports"),
    ("tcc.docx", "projects"),
    ("note_fiscal.pdf", "pdfs"),
    ("manual.pdf", "pdfs"),

    ("photo1.jpg", "photos_2024"),
    ("photo2.jpg", "trip"),

    ("rock1.mp3", "rock"),
    ("rock2.mp3", "rock"),
    ("pop1.mp3", "pop"),

    ("movie1.mp4", "movies"),
    ("movie2.mp4", "movies"),
]
```

```python
from anytree import Node, RenderTree

# criando os nodos da árvore
root = Node("root")

nodes = {
    "root": root
}

# construindo a árvore a partir da estrutura definida, onde cada tupla contém o nome do nodo e o nome do nodo pai
for name, parent_name in structure: # para cada tupla na estrutura, obtemos o nome do nodo e o nome do nodo pai
    parent_node = nodes[parent_name] # obtemos o nodo pai a partir do dicionário de nodos, usando o nome do nodo pai como chave
    nodes[name] = Node(name, parent=parent_node) # criamos um novo nodo com o nome do nodo atual e o associamos ao nodo pai, adicionando-o ao dicionário de nodos

for pre, fill, node in RenderTree(root):
    print(f"{pre}. {node.name}")

ext = input("Digite o tipo de arquivo que deseja buscar (ex: .pdf, .jpg, .mp3, .mp4, .docx, .xls): ")

count = {}
dfs(root, ext, count)

more_related_files = ["", 0]
# retornado um dicionario com a quantidade de arquivos encontrados em cada pasta, 
# verificamos qual pasta tem a maior quantidade de arquivos do tipo buscado
for count in count.items(): 
    if count[1] > more_related_files[1]:
        more_related_files = count

if more_related_files[1] == 0:
    print(f"Nenhum arquivo do tipo '{ext}' foi encontrado.")
else:
    print(
        f"A pasta recomendada é '{more_related_files[0]}', pois contém "
        f"{more_related_files[1]} arquivo(s) do tipo '{ext}'."
    )
```

### COMPLEXIDADE DO ALGORITMO 

Arvore + DFS: linear no número total de nodos, pois cada nodo é visitado uma vez. O bigO desse algotimo é O(n) pois varre todos os elementos da lista.
